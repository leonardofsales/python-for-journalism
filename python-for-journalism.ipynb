# Python for Data Journalists: Analyzing Money in Politics

These lines of code were written during the course "Python for Data Journalists: Analyzing Money in Politics", of the Knight Center for Journalism in Americas, held between June 12 and July 9, 2017.

The code reviews campaign financing in California in two databases: the first contains information on proposals and opposition and support committees of each proposal; the second gathers data on people and groups who have contributed funds to the campaigns. The code is written in Python/Pandas and Anaconda.

```python
import pandas as pd
```

```python
props = pd.read_csv("http://www.firstpythonnotebook.org/_static/committees.csv")
```

```python
props.head()
```

```python
props.info()
```

```python
contribs = pd.read_csv("http://www.firstpythonnotebook.org/_static/contributions.csv")
```

```python
prop_64 = props.prop_name == 'PROPOSITION 056 - CIGARETTE TAX TO FUND HEALTHCARE, TOBACCO USE PREVENTION, RESEARCH, AND LAW ENFORCEMENT. INITIATIVE CONSTITUTIONAL AMENDMENT AND STATUTE.'
prop = props[prop_64]
```

```python
prop.info()
```

```python
prop
```

```python
contribs.info()
```

```python
props.info()
```

```python
merged = pd.merge(prop, contribs, on="calaccess_committee_id")
```

```python
merged.head()
```

```python
merged.info()
```

```python
merged.amount.sum()
```

```python
merged.committee_position.value_counts()
```

```python
support = merged[merged.committee_position == 'SUPPORT']
oppose = merged[merged.committee_position == 'OPPOSE']
```

```python
oppose.amount.sum()
```

```python
support.amount.sum()
```

```python
support.amount.sum() / merged.amount.sum()
```

```python
merged.sort_values("amount", ascending=False)
```

```python
support.sort_values("amount", ascending=False).head()
```

```python
oppose.sort_values("amount", ascending=False).head()
```

```python
merged.groupby("committee_name_x").amount.sum()
```

```python
merged.groupby("committee_name_x").amount.sum().reset_index()
```

```python
merged.groupby("committee_name_x").amount.sum().reset_index().sort_values("amount", ascending=False)
```

```python
merged.groupby(["contributor_firstname", "contributor_lastname"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```

```python
merged.groupby(["contributor_firstname", "contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```
